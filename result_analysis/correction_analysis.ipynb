```python
from result_analysis import show_correction_performance, show_correction_similarity, show_correction_similarity_errors, compare_noise_types, compare_correction_similarity
import pandas as pd
```

```python
algorithms = ['PL', 'STC', 'CC', 'HLNC', 'OBNC', 'BE']
# algorithms =  ['OBNC', 'OBNC-remove-sensitive', 
#               'OBNC-optimize-demographic-parity-0', 'OBNC-optimize-demographic-parity-0.5',
#               'OBNC-fair-0', 'OBNC-fair-0.5']

experiments = ['phishing_having_IP_Address', 'bank_housing', 'monks1_attr3', 'biodeg_V25', 
                'credit_A1', 'sick_referral_source_other', 'vote_immigration', 'ads_local', 'churn_voice_mail_plan']

noise_types = ['random', 'flip', 'bias', 'balanced_bias']

nr = [i/10 for i in range(1,6)]
```
